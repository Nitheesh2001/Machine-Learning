```python
import pandas as pd
```

```python
df21 = pd.read_csv('titanic.csv')

# Extract cabin letter and number from non-null values
df21['Cabin'] = df21['Cabin'].str.extract(r'([A-Za-z])')
df21['Cabin_Number'] = df21['Cabin'].str.extract(r'(\d+)')
```

```python
df21
```

```python
cabin_counts = df21['Cabin'].value_counts(normalize=True)
missing_cabin_mask = df21['Cabin'].isnull()
df21.loc[missing_cabin_mask, 'Cabin'] = cabin_counts.index[0]
```

```python
median_cabin_number = df21['Cabin_Number'].median()
df21['Cabin_Number'].fillna(median_cabin_number, inplace=True)
```

```python
df21['Cabin'] = df21['Cabin'] + df21['Cabin_Number'].astype(str)
```

```python
df21.drop('Cabin_Number', axis=1, inplace=True)
```

```python
missing_cabin_count_after_fill = df21['Cabin'].isnull().sum()
print(f"Number of missing values in 'Cabin' column after fill: {missing_cabin_count_after_fill}")
```

```python
df21
```
